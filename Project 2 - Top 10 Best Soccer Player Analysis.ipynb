# FIFA Player Analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
players = pd.read_csv('fifa data.csv')
players.head(10)
```

### Top 10 Most Popular Player Nationality in FIFA

```python
nationality = players['nationality'].value_counts()[0:10]
nationality
```

```python
nationality = players['nationality'].value_counts()[0:10].keys()
nationality
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(players['nationality'].value_counts()[0:10].keys()),list(players['nationality'].value_counts()[0:10]))
plt.show()
```

### Wage vs Value Analysis of Top 10 Players in Fifa

```python
top_10 = players.iloc[0:10]
top_10
```

```python
top_10_wage = top_10[['short_name','wage_eur']]
top_10_value = top_10[['short_name','value_eur']]
```

```python
top_10_value = top_10_value.sort_values(by='value_eur',ascending=False)
top_10_value
```

```python
top_10_wage = top_10_wage.sort_values(by='wage_eur',ascending=False)
top_10_wage
```

```python
names = top_10.groupby('short_name')
short_name = [x for x, df in names]

wage=names.sum()['wage_eur']
value=names.sum()['value_eur']
```

```python
fig, ax = plt.subplots()
plt.title('Player Wage vs Value ')
ax2 = ax.twinx()
ax.bar(short_name, wage, color='dodgerblue')
ax2.plot(short_name, value, color='deeppink', lw=3, marker='1',mew=10)

ax.set_ylabel('Wage in Euros', color='dodgerblue')
ax2.set_ylabel('Value in Hundred Million Euros', color='deeppink')
ax.set_xticklabels(short_name,rotation='vertical',size=11)
plt.plot()
```

### Does Age Matter?

```python
age=names.sum()['age']
```

```python
fig, ax = plt.subplots()
plt.title('Player Age vs Value')
ax2 = ax.twinx()
ax.bar(short_name, age, color='limegreen')
ax2.plot(short_name, value, color='deeppink', lw=3, marker='1',mew=10)

ax.set_ylabel('Player Age', color='seagreen')
ax2.set_ylabel('Value in Hundred Million Euros', color='deeppink')
ax.set_xticklabels(short_name,rotation='vertical',size=11)
plt.plot()
```
